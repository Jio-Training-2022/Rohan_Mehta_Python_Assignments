```python
# Task
# 2. Print Table of 5, 6, 7, 8, 9, 10

for i in range(5,11):
    print()
    print('Table for {}'.format(i))
    for j in range(11):
        print('{} x {} = {}'.format(i,j,i*j))
```

```python
# Fizzbuzz Program

#FizzBuzz program

for i in range(1,101):
    if (i%3 == 0) and (i%5 == 0):
        print("fizzbuzz")
    elif (i%3 == 0):
        print("fizz")
    elif (i%5 == 0):
        print("buzz")
    else:
        print(i)
```
